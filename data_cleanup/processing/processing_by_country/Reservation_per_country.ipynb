# Pregled broja rezervacija po drzavi za hotel 1(id 0)

Prvo pronalazimo datoteku sa podatcima za hotel 1 i ucitavamo potrebne biblioteke.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

curr = Path.cwd()
curr = str(curr)
curr += "/country_data/no_of_reservations_per_country_hotel0.csv"
print(curr)
```

```python
df = pd.read_csv(curr)
df.info()
```

```python
dfn = df[df["Zemlja_gosta"] == "DEU"]
a = df.iloc[0].to_list()
a = a[2:]
b = dfn.columns.to_list()
b = b[2:]
#print(b)
c = []
d = []
ukupno = 0
for i in range(len(a)):
    if a[i] > 0:
        c.append(a[i])
        d.append(b[i])
    ukupno += a[i]
plt.figure(figsize=(70,6))
plt.plot(d,c)
plt.xticks(rotation=60)
print(ukupno)
```

Otvoriti cemo i bazu podataka za oba hotela gdje imamo i ukupan broj nocenja po rezervaciji

```python
curr = Path.cwd()
curr = str(curr)
curr += "/country_data/no_of_reservations_per_country.csv"
df2 = pd.read_csv(curr)
df2 = df2.drop('Unnamed: 0',axis = 1)
df2.info()
```

```python
a = list(df2["Ukupno"])
print(sum(a))
df2 = df2.sort_values(by = ["Ukupno"], ascending = False)
df2.head(n = 30)
```

```python
dfn = df2[df2["Zemlja_gosta"] == "FRA"]
a = dfn.iloc[0].to_list()
a = a[2:]
b = dfn.columns.to_list()
b = b[2:]
#print(b)
c = []
d = []
ukupno = 0
for i in range(len(a)):
    if a[i] > 50:
        c.append(a[i])
        d.append(b[i])
    ukupno += a[i]
plt.figure(figsize=(50,6))
for i in range(len(c)):
    c[i]/=ukupno
plt.plot(d,c)
plt.xticks(rotation=50)
print(ukupno)
```
